## Overview

In general terms Gradient means slope or slant of a surface. Gradient Descent aims to minimize the *cost function*. A function reaches its minimum value when the slope is equal to 0. Gradient descent is an iterative algorithm, that starts from a random point on a function and travels down its slope in steps until it reaches the minimum point of that function.

### Gradient Calculation from Cost function

>**Cost function:**
>
> ![](https://cdn.extras.talentsprint.com/aiml/Experiment_related_data/Images/cost_cal.jpg)

> Below equations shows gradient of Cost with respect to $θ_0$ and $θ_1$
>
> ![](https://cdn.extras.talentsprint.com/aiml/Experiment_related_data/Images/Gradient_calculation.jpg)
>
> **Note:** 'm' in above equations denote total number of samples, and is different from the 'm' present in `y=mx+c`

Let's visualize the Cost function first and then we will implement gradient of cost with respect to $θ_0$ and $θ_1$.

```python
# Visualize Cost function with varying values of theta_0 and theta_1

def cost_function(theta_0, theta_1):
    """
    returns cost function value for a particular theta_0 and theta_1 value
    cost = 1/2m * Σ(θ0 + θ1*x - y0)^2
    """
    theta = np.array([theta_0, theta_1])
    X_values = np.hstack((np.ones((len(x),1)), x))
    y_values = np.dot(X_values, theta)
    return np.sum((y_values - y)**2)/(2*len(x))


def get_cost_value_for_all_points(Ms, Cs):
    """
    Ms -> 'all theta_0 values'
    Cs -> 'all theta_1 values'
    returns cost function value for a particular theta_0 and theta_1 value
    """
    shape = Ms.shape
    global x, y
    Ms = Ms.ravel()       # flatten the array
    Cs = Cs.ravel()       # flatten the array
    result = []
    for i in range(len(Ms)):
        result.append(cost_function(Ms[i], Cs[i]))      # calculate cost function value for a particular theta_0 and theta_1 value
    return np.array(result).reshape(shape)



theta_0s = np.linspace(-20, 20, 200)               # 200 data points between -20 and 20 for theta_0
theta_1s = np.linspace(-20, 20, 200)               # 200 data points between -20 and 20 for theta_1

all_theta_0s, all_theta_1s = np.meshgrid(theta_0s, theta_1s)         # grid data points after intersection of theta_0 and theta_1 values
all_costs = get_cost_value_for_all_points(all_theta_0s, all_theta_1s)


# Plot cost function
# Ref. 3D plotting in Matplotlib :https://jakevdp.github.io/PythonDataScienceHandbook/04.12-three-dimensional-plotting.html

ax = plt.axes(projection='3d')
ax.plot_surface(all_theta_1s, all_theta_0s, all_costs, rstride=1, cstride=1, cmap='viridis')
ax.set_xlabel('θ1')
ax.set_ylabel('θ0')
ax.set_zlabel('Cost')
ax.set_title('Cost function');
```

#### Calculating gradient with respect to theta_0

```python
## ∂C/∂θ0 = 1/m * Σ(θ0 + θ1*x - y0)

(1/m) * np.sum((X.dot(theta) - y))    ## Ypredicted : X.dot(theta) , here y= original y
```

#### Calculating gradient with respect to theta_1

```python
## ∂C/∂θ1 = 1/m * Σ[(θ0 + θ1*x - y0)*x]

(1/m) * np.sum((X.dot(theta) - y) * x)
```

```python
X    ## shape (97,2)
```

```python
A = (X.dot(theta) - y)     ## Error e = (yp - yo)
A      ## shape (97,1)
```

### Combining both into one equation
* Here for simplicity only 3 rows of data (in both X and A matrix) have been shown, actually it has 'm' rows  containing all the data
* X is feature matrix
* **A** is the resulting matrix of error calculation i.e. result of eqn --> X.dot(theta)-Y    

![](https://cdn.extras.talentsprint.com/aiml/Experiment_related_data/Images/both_grad_in_one.jpg)


### Note the two elements of the resulting matrix:
* The first element is sum of each error, i.e. gradient with respect to theta_0, see the eqn above.
* The second element is the sum of the element-wise product of each error and  x, i.e. gradient with respect to theta_1.
* Thus if we, multiply X transpose with A we can get both gradients in one shot, as given below:

```python
(1/m) * (X.T).dot(X.dot(theta) - y)
```

#### Complete implementation

```python
## Theta initialization
np.random.seed(0)
theta = np.random.randn(2,1)  # first value=c, second value=m
print(theta)

n_iteration = 10000 ## number of iteration
a = 0.01

for i_iter in range(n_iteration):
    grad = (1/m) * (X.T).dot(X.dot(theta) - y) ##Gradient calculation
    theta = theta - a*grad

print(theta)         ## final updated theta after certain loop
```

### Including cost calculation for each iteration inside loop i.e. calculating cost for each updated theta:

```python
## Theta initialization
np.random.seed(0)
theta = np.random.randn(2,1)     # first value=c, second value=m
print(theta)

c = np.zeros((m,1))              # zero matrix initialization for storing cost values
n_iteration = 10000              # number of iterations
c = np.zeros((n_iteration,1))    # to store cost function value for each iteration
a = 0.001                        # Learning rate

for i_iter in range(n_iteration):
    grad = (1/m) * (X.T).dot(X.dot(theta) - y)          # Gradient calculation
    theta = theta - a*grad
    c[i_iter,0] = (1/(2*m)) * np.sum((np.square(X.dot(theta) - y)))

print(theta)          # final updated theta after certain loop
print(i_iter)
```

### Stochastic Gradient Descent Algorithm and its Implementation

```python
# Displaying shape of feature, target, and  first value of each array to feed into SGD

print(x.shape, x[0])        # feature
print(X.shape, X[0])        # feature in matrix form
print(y.shape,y[0])         # target
```

```python
n_epoch = 50
t0, t1 = 5, 500     # learning schedule hyperparameters

def learning_schedule(t):
    return t0/(t+t1)

theta = np.random.randn(2,1)       # assuming random theta_0 and theta_1 values
m = len(x)
cost = []


# Apply Stochastic Gradient Descent

for epoch in range(n_epoch):
    for i in range(m):       # In every epoch, the weights are updated exactly 'm' (the number of samples in the dataset) number of times.
        random_index = np.random.randint(m)
        Xi = X[random_index:random_index+1]      # A single sample is picked randomly, and weights are updated for each sample at a time.
        Yi = y[random_index:random_index+1]
        grad = Xi.T.dot((Xi.dot(theta)-Yi))      # Gradient calculation
        alpha = learning_schedule(epoch*m + i)   # Learning rate is updated according to the learning_schedule function.
        theta = theta - alpha*grad               # Update of weights
    cost.append((1/(2*m))*np.sum(np.square((X.dot(theta)-y))))   # Tracking of cost for each epoch

print(theta)
```

```python
# Displaying Cost vs Epoch
plt.plot(cost)
plt.ylabel('Cost')
plt.xlabel('Epoch')
plt.show()
```
